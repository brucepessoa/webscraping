### Webscraping do site da Glassdoor com lista de salários de Cientista de Dados

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python
# obtendo a url e a esrtutura de dados
resposta = requests.get('https://www.glassdoor.com.br/Sal%C3%A1rios/cientista-de-dados-sal%C3%A1rio-SRCH_KO0,18.htm')
html = resposta.text

#arrumando o html hierarquizado
html_estruturado = BeautifulSoup(html, 'html.parser')
```

```python
# buscando os nomes das empresas
lista_empresas = html_estruturado.find_all('h3',
                {'data-test': re.compile('salaries-list-item-.*-employer-name')})

for empresa in lista_empresas:
  nome_empresa = empresa.find('a').text
  print(nome_empresa)
```

```python
# buscando os salários
lista_salarios = html_estruturado.find_all('div',
                {'data-test': re.compile('salaries-list-item-.*-salary-info')})

for salario in lista_salarios:
  str_salario = salario.contents[0].text
  str_salario = str_salario.replace('R$','').replace('\xa0','').replace('.','')
  print(str_salario)
```

```python
# listando empresa - salário
for empresa, salario in zip(lista_empresas, lista_salarios):
  nome_empresa = empresa.find('a').text
  str_salario = salario.contents[0].text
  str_salario = str_salario.replace('R$','').replace('\xa0','').replace('.','')

  print(f'{nome_empresa} - {str_salario}')
```

```python
# criando o dataframe 
lista_todos_salarios = []

for empresa, salario in zip(lista_empresas, lista_salarios):
  nome_empresa = empresa.find('a').text
  str_salario = salario.contents[0].text
  str_salario = str_salario.replace('R$','').replace('\xa0','').replace('.','')
 
  # adiciona na lista uma tupla com empresa / salario
  lista_todos_salarios.append((nome_empresa, str_salario))


# transformando em dataframe
df_salarios = pd.DataFrame(lista_todos_salarios)

# renomeando as colunas
df_salarios = pd.DataFrame(lista_todos_salarios, columns=['empresa','salario'])

# transofrmando salário de string para float
df_salarios['salario'] = df_salarios['salario'].astype(np.float32)
```
